```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)
```

```python
#Opening the HomePage Yatra.com
url='https://www.yatra.com/'
driver.get(url)

time.sleep(2)
```

```python
#User have to enter source station ,destination, and date of flight
#Then wse have to click on search botton.
search_button=driver.find_element_by_xpath('//div[@class="ripple-parent search-height demo-icon icon-go"]//input')
search_button.click()

time.sleep(5)
```

```python
#Extracting tag that contains airline_name
airline_name=driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
airline_name
```

```python
len(airline_name)
```

```python
Airline_name=[]
for i in range(0,160):
    try:
        Airline_name.append(airline_name[i].text)
    except NoSuchElementException:
        Airline_name.append('-')
Airline_name 
```

```python
#Extracting source tag
source_station=driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
source_station
```
